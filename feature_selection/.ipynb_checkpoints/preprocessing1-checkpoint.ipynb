# Data preprocessing

### 1. Remove "PassengerId", "Ticket". Passenger id is just serial number, ticket is distinct number both doesnot convey any meaning, also drop "Cabin", "Embarked" columns 

```python
df.drop(['PassengerId', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
```

```python
df.head()
```

### 2. Map male, female to integer values, male=0, female=1

```python
gender_mapper = {"male": 0, "female": 1}
df['Sex'].replace(gender_mapper, inplace=True)
```

```python
df.head()
```

### 3. Change the "Name" column to someting meaningful. [Mr., Mrs., Miss.] = 0, other titles give 1. Better create one more column called "title",  finally drop the "Name" column

```python
df['title'] = df['Name'].apply(lambda x: x.split(',')[1].strip().split(' ')[0])
```

```python
df['title'] = [0 if x in ['Mr.', 'Mrs.', 'Miss.'] else 1 for x in df.title]
```

```python
df.drop(['Name'], axis=1, inplace=True)
```

```python
df.head()
```

### 4. Fill the na values with mean values for the "Age" column

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.head()
```

Now all the columns are in integer format, we can use this dataframe now to train the models.
