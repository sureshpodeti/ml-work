```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(2, 50, 1)
```

```python
Y = X**2
```

plt.subplots() generates the plots like grids called subplot, if we dont give any arguments by default it generates 1x1 subplot.

if given plt.subplots(m,n), it generates mxn subplots

fig is the entire figure. aux is the axis which give access to the each subplot. 

we can access the each subplot by giving indexes. like aux[0,0] first the first. 

```python
fig, aux = plt.subplots(2,2)
aux[0,1].plot(X,Y)
plt.show()
```

```python
fig, aux = plt.subplots()
aux.plot(X,Y)
aux.set_xlabel("X Label")
aux.set_ylabel("Y Label")
aux.set_title("Relationship b/w X and Y")
plt.show()
```
