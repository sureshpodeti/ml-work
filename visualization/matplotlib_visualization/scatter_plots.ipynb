```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(2, 50,2)
Y = X**2
```

```python
plt.figure(1)
plt.subplot(111)
plt.scatter(X,Y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

Same thing can be done as follows:

```python
fig, aux = plt.subplots()
plt.scatter(X,Y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
```

```python
Y = np.reshape(iris.target, (X.shape[0],1))
```

```python
data = np.hstack([X,Y])
```

```python
plt.plot(data[:, 0][data[:, -1]==0], data[:, 1][data[:, -1]==0], 'ro')
plt.plot(data[:, 0][data[:, -1]==1], data[:,1][data[:, -1]==1], 'bo')
plt.plot(data[:, 0][data[:, -1]==2], data[:,1][data[:, -1]==2], 'go')
plt.xlabel("Sepal length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
X = np.arange(2, 50,2)
Y = X**2
```

```python
fig, aux = plt.subplots()
aux.scatter(X, Y)
aux.set_xlabel("X Axis")
aux.set_ylabel("Y Axis")
aux.set_title("Relationship b/w X and Y")
plt.show()
```
