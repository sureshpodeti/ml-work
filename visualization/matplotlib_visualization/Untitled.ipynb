```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([1,2,3,4])
```

if Y value is not provided python take Y values from 0 to len(X)-1

```python
plt.plot(X, X, 'ro', X, X**2, 'b-', X, X**3, 'gs', X, 2*X+8, 'y^')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

plt.axis([xmin, xmax, ymin, ymax]) Defines the range of X,Y axes

```python
data = {'a': np.arange(50),
        'c': np.arange(10, 60),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
```

## categorical variables plotting

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.figure(figsize=(9,3))
```

```python
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle("Categorical Plotting")
plt.show()
```
